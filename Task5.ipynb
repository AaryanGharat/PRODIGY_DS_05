```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import plotly.express as px
import plotly.offline as pyo
```

```python
pwd = os.getcwd().replace("\\", "/")
filepath=os.path.join(pwd,"TASK5 DATASET/US_Accidents_March23.csv")
df=pd.read_csv(filepath)
df
```

```python
df['Start_Time'] = pd.to_datetime(df['Start_Time'])
df['End_Time'] = pd.to_datetime(df['End_Time'])
df['Hour'] = df['Start_Time'].dt.hour
df['DayOfWeek'] = df['Start_Time'].dt.dayofweek
df['Month'] = df['Start_Time'].dt.month
```

**Accident Frequency by Month**

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Month', data=df, palette='viridis')
plt.title('Accident Frequency by Month')
plt.xlabel('Month')
plt.ylabel('Accident Count')
plt.show()
```

**Accidents by Road Conditions**

```python
road_condition_columns = ['Amenity', 'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit', 'Railway', 'Roundabout', 'Station', 'Stop', 'Traffic_Calming', 'Traffic_Signal', 'Turning_Loop']
plt.figure(figsize=(14, 6))
road_condition_counts = df[road_condition_columns].sum().sort_values(ascending=False)
sns.barplot(x=road_condition_counts, y=road_condition_counts.index, palette='plasma')
plt.title('Accidents by Road Conditions')
plt.xlabel('Accident Count')
plt.ylabel('Road Conditions')
plt.show()
```

**Weather condition VS Accidents**

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(16, 12))
sns.histplot(df, x='Temperature(F)', hue='Severity', multiple="stack", bins=30, ax=axes[0, 0])
axes[0, 0].set_title('Temperature vs. Severity')
sns.histplot(df, x='Wind_Speed(mph)', hue='Severity', multiple="stack", bins=30, ax=axes[0, 1])
axes[0, 1].set_title('Wind Speed vs. Severity')
sns.histplot(df, x='Visibility(mi)', hue='Severity', multiple="stack", bins=30, ax=axes[1, 0])
axes[1, 0].set_title('Visibility vs. Severity')
sns.histplot(df, x='Precipitation(in)', hue='Severity', multiple="stack", bins=30, ax=axes[1, 1])
axes[1, 1].set_title('Precipitation vs. Severity')
plt.tight_layout()
plt.show()
```

**Hotspot of Accident**

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Start_Lng', y='Start_Lat', data=df, alpha=0.3, s=5)
plt.title('Accident Hotspots')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
